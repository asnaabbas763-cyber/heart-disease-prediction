## 1.1 Problem Statement

Cardiovascular disease (CVD) remains the **leading cause of death globally**, accounting for approximately 17.9 million deaths per year according to the World Health Organization (WHO). Early and accurate diagnosis is critical — yet traditional clinical assessment can be time-consuming, costly, and dependent on specialist availability.

This project addresses the following core question:

> **Can we reliably predict the presence or absence of heart disease in a patient using a set of 13 routine clinical measurements, and which machine learning model performs best for this task?**

Specifically, we aim to:
1. Build a **binary classification model** that predicts whether a patient has heart disease (target = 1) or not (target = 0).
2. Compare the performance of **K-Nearest Neighbours (KNN)**, **Logistic Regression**, and **Gradient Boosting**.
3. Identify the **most diagnostically significant clinical features** using Mutual Information and Chi-Square feature selection.
4. Deploy the best-performing model as an **interactive web application** for clinical decision support.

## 1.2 Clinical Background

Heart disease is an umbrella term covering several conditions affecting the heart's structure and function. The most common form is **coronary artery disease (CAD)**, caused by the narrowing or blockage of coronary arteries due to atherosclerosis (plaque build-up).

### Key Risk Factors (aligned with our dataset features):

| Clinical Feature | Medical Relevance |
|---|---|
| **Age** | Risk increases significantly after age 45 (men) and 55 (women) |
| **Sex** | Men have higher risk; post-menopausal women catch up |
| **Chest Pain Type (cp)** | Typical angina is a strong indicator of CAD |
| **Resting Blood Pressure (trestbps)** | Hypertension strains the heart and damages arteries |
| **Cholesterol (chol)** | High LDL cholesterol accelerates plaque build-up |
| **Fasting Blood Sugar (fbs)** | Diabetes dramatically increases CVD risk |
| **Resting ECG (restecg)** | ST-T wave abnormalities indicate myocardial ischemia |
| **Max Heart Rate (thalach)** | Lower exercise capacity often signals cardiac dysfunction |
| **Exercise-induced Angina (exang)** | Chest pain during exertion is a classic ischemia sign |
| **ST Depression (oldpeak)** | Exercise-induced ST depression linked to ischemia severity |
| **Slope of Peak ST (slope)** | Down-sloping ST associated with CAD |
| **Major Vessels coloured (ca)** | More blocked vessels = higher disease severity |
| **Thalassemia (thal)** | Reversible defects suggest ischemia; fixed defects indicate scarring |

## 1.3 Literature Review

Machine learning has been increasingly applied to cardiac diagnosis over the past two decades. Below is a summary of key studies that motivate our methodology:

---

### Study 1: Detrano et al. (1989) — The Cleveland Dataset Origin
- **Paper:** *International application of a new probability algorithm for the diagnosis of coronary artery disease*
- **Journal:** American Journal of Cardiology
- **Key Findings:** The Cleveland dataset was originally assembled to evaluate an angiographic diagnosis model using 303 patients. The 14-attribute version (13 features + target) became the gold-standard benchmark for cardiac ML research.
- **Relevance:** Our dataset is derived directly from this study. The authors used logistic regression — we extend this with modern ensemble methods.

---

### Study 2: Bashir et al. (2019) — KNN for Heart Disease
- **Paper:** *Applying K-Nearest Neighbour to Heart Disease Diagnosis*  
- **Key Findings:** KNN achieved ~83% accuracy on the Cleveland dataset with k=5. The study noted KNN is sensitive to feature scaling and recommended StandardScaler normalisation — which we adopt.
- **Relevance:** Validates KNN as a competitive baseline; informs our hyperparameter (k=5) starting point.

---

### Study 3: Mohan et al. (2019) — Hybrid Random Forest with Feature Selection
- **Paper:** *Effective Heart Disease Prediction Using Hybrid Machine Learning Techniques*
- **Journal:** IEEE Access
- **Key Findings:** Combined feature selection (Mutual Information + Chi-Square) with ensemble classifiers. MI + chi-square hybrid selection improved accuracy by 4–6% over using raw features. `cp`, `thalach`, `ca`, and `thal` were consistently ranked as top predictors.
- **Relevance:** Directly motivates our dual feature-selection strategy (MI + Chi-Square).

---

### Study 4: Shah et al. (2020) — Gradient Boosting for CVD
- **Paper:** *Heart Disease Prediction using Machine Learning Techniques*
- **Key Findings:** Gradient Boosting (XGBoost variant) achieved up to 92% accuracy on the Cleveland dataset, outperforming SVM and Naive Bayes. Feature importance from GB models aligned with clinical literature on risk factors.
- **Relevance:** Justifies inclusion of Gradient Boosting as a high-complexity model in our comparison.

---

### Study 5: Weng et al. (2017) — NHS Cardiovascular Risk Prediction
- **Paper:** *Can machine-learning improve cardiovascular risk prediction using routine clinical data?*
- **Journal:** PLOS ONE
- **Key Findings:** ML models (particularly Random Forest and Neural Networks) outperformed the gold-standard QRISK2 tool on NHS patient data, correctly identifying 7.6% more patients at risk.
- **Relevance:** Demonstrates real-world clinical deployment potential — motivates our Streamlit deployment goal.

---

### Summary of Literature Gap
While many studies compare models individually, fewer provide:
1. A **reproducible end-to-end pipeline** from raw data to deployed web app.
2. **Visual decision boundaries** for interpretability alongside quantitative metrics.
3. **Combined Mutual Information and Chi-Square selection** with explicit feature ranking visualisations.

Our project addresses all three gaps.

## 1.5 Ethical Considerations

As this project involves clinical data and medical predictions, we acknowledge the following:

- **Data Privacy:** The Cleveland dataset is fully anonymised. No personally identifiable information (PII) is present.
- **Model Limitations:** Our model is a research/educational tool — it is **NOT** a substitute for professional medical diagnosis. False negatives (missed disease cases) carry higher clinical risk than false positives.
- **Class Imbalance Awareness:** The dataset has a slight imbalance (165 disease vs 138 no-disease). We use stratified splits and report precision/recall alongside accuracy to avoid misleading performance claims.
- **Bias & Fairness:** The Cleveland dataset is primarily from a single US institution. Generalisability to broader populations (different ethnicities, geographies) is limited and should be studied before any real-world deployment.
